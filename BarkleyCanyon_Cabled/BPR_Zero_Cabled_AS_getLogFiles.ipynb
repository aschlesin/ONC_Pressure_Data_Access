```python
from onc import ONC
import pandas as pd
import logging
import os
from dotenv import load_dotenv
load_dotenv()

onc = ONC(token=os.getenv('ONC_TOKEN'))
```

```python
deviceCode = 'RBRQUARTZ3BPRZERO207223' # at Barkley Canyon
```

```python
import requests_cache

s = requests_cache.CachedSession('requests_cache')
```

## get logFiles from Archive from ONC OpenAPI

```python
data = onc.getArchivefileByDevice({
    'deviceCode':deviceCode,
    'dateFrom': '2022-04-05T00:00:00.000Z',
    'dateTo': '2022-04-06T00:00:00.000Z',
})
```

```python
data['files']
```

### Next step will create a  folder called output and downloads the files into

```python
for file in data['files']:
    if not os.path.exists ('output/'+file):
        onc.downloadArchivefile(filename=file)
```
