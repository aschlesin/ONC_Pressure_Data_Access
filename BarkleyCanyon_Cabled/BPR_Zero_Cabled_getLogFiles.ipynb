## BPR|Zero Cabled — Download Archive Log Files

Download raw archive (log) files from the ONC archive for the cabled **RBR Quartz BPR|Zero** at Barkley Canyon.

**Device:** `RBRQUARTZ3BPRZERO207223`

**Setup:** Store your ONC API token in a `.env` file in this directory as `ONC_TOKEN=<your_token>`.  
Tokens are available from your [ONC account page](https://data.oceannetworks.ca/Profile).

```python
import os

from onc import ONC
from dotenv import load_dotenv

load_dotenv()

# Initialise the ONC Python client library with the API token
onc = ONC(token=os.getenv('ONC_TOKEN'))
```

```python
# RBR Quartz BPR|Zero cabled at Barkley Canyon
deviceCode = 'RBRQUARTZ3BPRZERO207223'
```

### List available archive files

Query the ONC archive for log files produced by the device within the specified date range.

```python
data = onc.getArchivefileByDevice({
    'deviceCode': deviceCode,
    'dateFrom': '2022-04-05T00:00:00.000Z',
    'dateTo': '2022-04-06T00:00:00.000Z',
})
```

```python
# Display the list of available archive file names
data['files']
```

### Download archive files

Download each file to the local `output/` directory (created automatically by the ONC client). Files already present are skipped to avoid redundant downloads.

```python
for file in data['files']:
    if not os.path.exists('output/' + file):
        onc.downloadArchivefile(filename=file)
```
